```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

x = np.array([128, 256, 512, 1024, 2048, 4096, 8192, 16384, 32768, 65536, 131072, 262144])
y1 = np.array([1.208828, 1.190433, 1.199451, 1.194965, 1.217502, 1.205639, 1.20461, 1.198474, 1.202379, 1.20587, 1.202841, 1.202273])
y4 = np.array([28.515652, 47.17791, 78.498199, 130.052948, 216.677155, 361.409027, 604.380127, 1007.744995, 1687.083618, 2827.974609, 4736.601929, 7966.826172])

fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(121)
ax.set_title('Dimension 0')
ax.scatter(x, y1, color='blue', label='Sine wave')
ax.set_ylim([1.1, 2.1])

m1, b1 = np.polyfit(x, y1, 1)
ax.plot(x, m1*x + b1)
print("y = " + str(round(m1,10)) + "x + " + str(round(b1,3)))
```

```python
y2 = np.array([7.700475, 10.814993, 14.918116, 21.15583, 29.666779, 41.711098, 58.976452, 83.117874, 117.520584, 165.978821, 234.7371156, 331.4753013])

fig = plt.figure(figsize=(12, 6))
ax2 = fig.add_subplot(121)
ax2.set_title('Dimension 2')
ax2.scatter(x, y2, color='blue', label='Sine wave')
ax2.set_ylim([7, 340])

def exp_fit(x,a,b,c):
    y = a*x**b + c
    return y

fit = curve_fit(exp_fit, x, y2)
fit_eq = fit[0][0]*x**fit[0][1] + fit[0][2]
ax2.plot(x, fit_eq)
print(str(round(fit[0][0],3)) + "x^" + str(round(fit[0][1],3)) + " + " + str(round(fit[0][2],3)))
```
